```python
px = []
py = []

with open('particles_coordinates.txt') as f:
    f.readline(); f.readline(); f.readline();
    for i in range(40**2):
        temp = f.readline()[:-1]
        temp = temp.split(',')
        px.append(float(temp[0]))
        py.append(float(temp[1]))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,8))
axpx = fig.add_subplot(121); axpx.set_title('px'); 
axpy = fig.add_subplot(122); axpy.set_title('py');


axpx.hist(px)
axpy.hist(py)

plt.show()
#fig.savefig('distribution of px and py.png')
```
