```python
import numpy as np
Wp = []
Wa = []

with open("energy_values.txt", 'r') as f:
    TIME_dt = f.readline()[:-1]
    TIME = (int)(TIME_dt.split(' ')[0])
    dt = (float)(TIME_dt.split(' ')[1])
    W0 = (float)(f.readline())
    for t in range(TIME):
        Wp_Wa = f.readline()[:-1]
        Wp.append(float(Wp_Wa.split(' ')[0]))
        Wa.append(float(Wp_Wa.split(' ')[1]))
        
Wp = np.asarray(Wp)
Wa = np.asarray(Wa)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import datetime
import math

%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211, xlim=(0, TIME)); ax1.set_title(r"$[(W_a - W_0)/W_0](\hat{t} \cdot w_p)$")
ax2 = fig.add_subplot(212, xlim=(0, TIME)); ax2.set_title(r"$[(W_p - W_0)/W_0](\hat{t} \cdot w_p)$")

t = np.asarray(range(TIME))
dWa = [(w-W0)/W0 for w in Wa]
dWp = [(wp-W0)/W0 for wp in Wp]

ax1.plot(t, dWa, color = 'blue')
ax2.plot(t, dWp, color = 'orange')

now = datetime.datetime.now()
fig.savefig("evolution of heated plasma, ref. boundaries (" + now.strftime('%Y%m%d, %H_%M') + ").png")

plt.show()
```
