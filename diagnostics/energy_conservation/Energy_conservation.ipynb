```python
I = []

with open("energy_values.txt", 'r') as f:
    TIME = (int)(f.readline())
    I0 = (float)(f.readline())
    for t in range(TIME)[:-3]:
        I.append(float(f.readline()[:-1]))
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math

%matplotlib inline

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, xlim=(0, TIME)); ax.set_title("(energy - energy0)/energy0_(t)")

dI = [(i-I0)/I0 for i in I]


ax.plot(range(TIME)[:1003-3], dI[:1000], color = 'blue')

#fig.savefig("acceleration of elecrton in electrostatic field 20201121.png")
plt.show()
```
