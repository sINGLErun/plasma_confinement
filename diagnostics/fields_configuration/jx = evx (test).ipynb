```python
#ev - 
j = []

with open("jx.txt", 'r') as f:
    TIME = (int)(f.readline())
    for t in range(TIME):
        j.append(float(f.readline()))
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.axes_grid1 import make_axes_locatable
from statistics import mean

%matplotlib inline

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, xlim=(0, TIME))#, ylim=(0.445, 0.450))

#dI = [(i-I0)/I0 for i in I]


ax.plot(range(TIME), j)

fig.savefig("jx = evx.png")
```
