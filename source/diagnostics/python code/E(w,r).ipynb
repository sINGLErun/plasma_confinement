```python
field_at = []
with open("field_at.txt", 'r') as f:
    TIME_dt = f.readline()[:-1]
    TIME = (int)(TIME_dt.split(' ')[0])
    dt = (float)(TIME_dt.split(' ')[1])
    for t in range(TIME):
        field_at.append((float)(f.readline()))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.fftpack import rfft, rfftfreq
import datetime

f_t = np.asarray(field_at[:])
    
f_w = rfft(f_t)
w = rfftfreq(TIME, dt)*2*np.pi

fig = plt.figure(figsize=(10,12))
ax1 = fig.add_subplot(211); ax1.set_title('Ex(t)')
ax2 = fig.add_subplot(212); ax2.set_title('Ex(w)')

f_t_ = ax1.plot(range(TIME), field_at)
f_w_ = ax2.plot(w[:250], (np.abs(f_w)/len(f_w))[:250])

now = datetime.datetime.now()
#fig.savefig("manual grid.png")
```
