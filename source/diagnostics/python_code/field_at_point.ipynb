```python
field_at = []
with open("Bz.txt", 'r') as f:
    TIME_dt_DTS = f.readline().split(' ')
    TIME = (int)(TIME_dt_DTS[0])
    dt = (float)(TIME_dt_DTS[1])
    DTS = (int)(TIME_dt_DTS[2])
    for t in range(int(TIME/DTS)):
        field_at.append((float)(f.readline()))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.fftpack import rfft, rfftfreq
import datetime

f_t = np.asarray(field_at[:])
    
f_w = rfft(f_t)
w = rfftfreq(int(TIME/DTS), dt)*2*np.pi

fig = plt.figure(figsize=(10,12))
ax1 = fig.add_subplot(211); ax1.set_title('$B_z\ (x=Nx/2,\ y=Ny/2,\ t) $')
ax2 = fig.add_subplot(212); ax2.set_title('$B_z\ (x=Nx/2,\ y=Ny/2,\ t[5e3:])$')

f_t_ = ax1.plot(range(0, TIME, DTS), field_at)
#f_w_ = ax2.plot(w, (np.abs(f_w)/len(f_w)))
f_t__ = ax1.plot(range(0, TIME, DTS)[100:], field_at[100:], color='red')
f_t___ = ax2.plot(range(0, TIME, DTS)[100:], field_at[100:], color='red')


now = datetime.datetime.now()
#fig.savefig("Bz(0.5'SX, 0.5'SY, t).png")
```
